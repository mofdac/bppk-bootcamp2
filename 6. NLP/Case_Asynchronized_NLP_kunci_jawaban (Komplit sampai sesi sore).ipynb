```python
#import libraries yang diperlukan
import pandas as pd
import re

# list tanda baca 
from string import punctuation

# stemmer bahasa indonesia
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# stopwords bahasa indonesia
from nltk.corpus import stopwords
stop = list(stopwords.words('indonesian'))


pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)
```

### LOAD DATASET 

```python
dataset = pd.read_csv('dataset_tweet_sentiment_opini_film.csv')
dataset.info()
```

```python
dataset.head(5)
```

### Text Normalization
#### Lowercasing

```python
# isikan stepnya disini, gunakan kolom Text Tweet dan simpan di kolom baru Text Lower
dataset['Text Lower'] = dataset['Text Tweet'].str.lower()
```

```python
dataset.head(5)
```

### Text Cleansing
#### handle tweet unique parts

```python
# menghilangkan/menyatukan # menjadi satu tokens (HASHTAG)
#baik untuk #dengan kata atau #dengan angka

dataset['Tweet Clean'] = dataset['Text Lower'].replace(to_replace=r'#(\w|\d)+', value='HASHTAG', regex=True)
```

```python
#menghilangkan/menyatukan http atau https menjadi satu tokens (URL)
#pengerjaan masih di kolom yang sama (Tweet Clean)

dataset['Tweet Clean'] = dataset['Tweet Clean'].replace(to_replace=r'(http|https)\S+', value='URL', regex=True)
```

```python
# menghilangkan/menyatukan @ menjadi satu tokens (TAGGING)
# baik untuk @dengan kata atau @dengan angka
# pengerjaan masih di kolom yang sama (Tweet Clean)

dataset['Tweet Clean'] = dataset['Tweet Clean'].replace(to_replace=r'@(\w|\d)+', value='TAGGING', regex=True)
```

```python
dataset['Tweet Clean'] = dataset['Tweet Clean'].str.strip()
dataset.head()
```

#### Remove Number

```python
# menghapus angka pada tweet (menjadi empty space ''), gunakan kolom Tweet Clean dan simpan di kolom yang sama
# gunakan regex \d untuk mencari angka

dataset['Tweet Clean']  = dataset['Tweet Clean'].str.replace('\d+', ' ')
```

```python
dataset['Tweet Clean'] = dataset['Tweet Clean'].str.strip()
dataset.tail()
```

#### Remove Punctuation

```python
#count punctuation before implementation
dataset['Tweet Clean'].map(lambda v: any(char in v for char in punctuation)).sum()
```

```python
#remove_puctuations functions
def remove_punctuations(text):
    # tuliskan algoritma remove punctuationnya
    for punct in punctuation:
        text = text.replace(punct, '')
    return text

# apply fungsinya pada kolom tweet clean dan simpan pada kolom baru Punct Removed
dataset['Punct Removed'] = dataset['Tweet Clean'].apply(remove_punctuations)
```

```python
dataset['Punct Removed'] = dataset['Punct Removed'].str.strip()
# cek count lagi
print('sisa punct:',dataset['Punct Removed'].map(lambda v: any(char in v for char in punctuation)).sum())

dataset.head()
```

#### Remove Slang Words

```python
# load file slang word 
normalized_slang = pd.read_excel('Slang_word.xlsx')
normalized_slang.head()
```

```python
def remove_slang(text):
    # tuliskan algoritma untuk mereplace kata tidak baku menjadi kata baku
    list_tidak_baku = normalized_slang.Tidak_baku.to_list()
    list_baku = normalized_slang.Baku.to_list()
    for i in range(normalized_slang.shape[0]):
        if list_tidak_baku[i] in text:
            text = re.sub(r'\b(?:'+list_tidak_baku[i]+r')\b',list_baku[i],text)
    return text

# apply fungsinya pada kolom Punct Removed dan simpan pada kolom baru Slang Removed
dataset['Slang Removed'] = dataset['Punct Removed'].apply(remove_slang)
```

```python
dataset['Slang Removed'] = dataset['Slang Removed'].str.strip()
dataset.head()
```

#### Remove Stopwords

```python
# keluarkan kata tidak dalam list stopwords

print(len(stop))
stop.remove('tidak')
print(len(stop))
```

```python
def remove_stopword(text):
# tulis di sini codenya, gunakan kolom Slang Removed dan simpan di kolom baru Stopwords Removed
    for i in range(len(stop)):
        if stop[i] in text:
            text = re.sub(r'\b(?:'+stop[i]+r')\b','',text)
            text = text.replace('  ',' ')
    return text

dataset['Stopwords Removed'] = dataset['Slang Removed'].apply(remove_stopword)
```

```python
dataset['Stopwords Removed'] = dataset['Stopwords Removed'].str.strip()
dataset.head()
```

#### Stemming (pengembalian kata ke kata dasar)

```python
# load stemmer dari package sastrawi yang telah di import 
stemmer = StemmerFactory().create_stemmer()
```

```python
def stem_sentences(sentence):
#tulis step-step pengaplikasian stemmer di atas
    tokens = sentence.split()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return " ".join(stemmed_tokens)

# apply fungsi diatas
dataset['Stem tweet'] = dataset['Stopwords Removed'].apply(stem_sentences)
```

```python
dataset['Stem tweet'] = dataset['Stem tweet'].str.strip()
dataset.tail()
```

#### Remove short words (less than 3 characters)

```python
# tulis step untuk meremove kata-kata yang singkat 1-2 character
# gunakan kolom Stem tweet dan simpan pada kolom baru Short-Tokens

def remove_short(text):
    tokens = text.split()
    for token in tokens:
        if len(token) < 3:
            tokens.remove(token)
    return ' '.join(tokens)

dataset['Short-Tokens'] = dataset['Stem tweet'].apply(remove_short)
```

```python
dataset['Short-Tokens'] = dataset['Short-Tokens'].str.strip()
dataset.head()
```

## Predictive Modeling
#### Labeling Target Variable

```python
#merubah label y, dalam hal ini sentimen negatif dan positif ke dalam bentuk biner
#bisa menggunakan berbagai macam metode

dataset['target'] = dataset['Sentiment'].map({'negative':0,'positive':1})
dataset['target'].value_counts()
```

```python
dataset.head(5)
```

#### Train Test Split

```python
#membagi dataset ke dalam train dan test
#pastikan sudah terinstall library train_test_split
from sklearn.model_selection import train_test_split

#perhatikan parameter yang digunakan saat membagi dataset
x_train, x_test, y_train, y_test = train_test_split(dataset['Short-Tokens'],
                                                    dataset['target'],
                                                    shuffle = True,
                                                    stratify=dataset['target'],
                                                    test_size = 0.2,
                                                    random_state = 1717
                                                   )

#dari 200 baris dataset
#terbagi menjadi 160 training set dan 40 testing set
#hal ini bergantung pada parameter test_size
print('Baris Training Set \t:', str(len(x_train)))
print('Baris Test Set \t\t:', str(len(x_test)))

#distribusi dari masing-masing training set dan testing set atas variable y
#dalam hal ini sentimen negatif dan positif
print('\nDistribusi target variable training set:\n', y_train.value_counts())
print('\nDistribusi target variable test set:\n', y_test.value_counts())
```

### Feature Generation
#### TF-IDF Vectorizer

```python
#mengimplementasikan TF-IDF
#pastikan sudah terinstall library TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

#memanggil library
vectorizer = TfidfVectorizer()

#melakukan fitting ke train 
vectorizer.fit(x_train)

#melakukan transformasi atas x_train ke dalam bentuk vektor
x_train_vec = vectorizer.transform(x_train)

#jumlah keseluruhan fitur yang berhasil di ekstraksi atas implementasi tf-idf
len(vectorizer.get_feature_names())
```

#### Predictive Model

```python
#menggunakan algoritma MultinomialNB untuk melakukan prediksi
#pastikan sudah terinstall library MultinomialNB
from sklearn.naive_bayes import MultinomialNB

#memanggil library
model = MultinomialNB()

#melakukan fitting atas training set, dalam hal ini x_train_vec (xtrain dalam bentuk vektor) dan y_train
model.fit(x_train_vec, y_train)
```

#### Evaluation Metric

```python
#mengukur kemampuan dari model dalam melakukan prediksi
#menggunakan accuracy_scor dan confusion_matrix untuk melihat TP, TN, FP, dan FN
#pastikan sudah terinstall library accuracy_score dan confusion_matrix
from sklearn.metrics import accuracy_score, confusion_matrix

#melakukan transformasi atas x_test ke dalam bentuk vektor
x_test_vec = vectorizer.transform(x_test)

#melakukan prediksi atas x_test_vec (x_test dalam bentuk vektor)
prediction = model.predict(x_test_vec)

#melihat accuracy_score
print('Score akurasi model:',str(accuracy_score(y_test,prediction)))

#melihat confusion_score
confusion_matrix(y_test, prediction)
```

```python
pd.DataFrame({'text':x_test, 'true label':y_test,'prediction':prediction})
```

#### Hyperparameter Tuning

```python
#melakukan hyperparameter tuning, artinya mengubah parameter dalam algoritma
#dalam rangka memperoleh hasil terbaik, dalam hal ini accuracy_score dan confusion_matrix
#disini menggunakan library GridSearchCV
#artinya seluruh parameter yang kita masukan ke dalam param_grid, akan dicoba satu-persatu
#pastikan sudah terinstall library GridSearchCV
from sklearn.model_selection import GridSearchCV

#range dari setting parameter yang akan kita coba
param_grid = {'alpha': [0.01, 0.1, 1, 10]}
```

```python
#perhatikan seting dari masing-masing parameter di GridSearchCV
g_search = GridSearchCV(estimator = model, param_grid = param_grid, cv = 3, n_jobs = 1, verbose = 0, return_train_score=True)
```

```python
#melakukan fitting atas training set
g_search.fit(x_train_vec, y_train)
```

```python
#memperoleh parameter terbaik
print(g_search.best_params_)
```

```python
#menggunakan parameter terbaik ke dalam model
best_model = MultinomialNB(0.1)
best_model.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

x_test_vec = vectorizer.transform(x_test)

prediction = best_model.predict(x_test_vec)

print('Score akurasi model:',str(accuracy_score(y_test,prediction)))

confusion_matrix(y_test, prediction)
```
