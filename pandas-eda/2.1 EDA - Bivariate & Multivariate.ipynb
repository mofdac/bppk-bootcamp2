### Bivariate Analysis

Analisis Bivariate merupakan analisis yang dilakukan terhadap 2 variabel untuk mengetahui hubungan antara kedua variabel tersebut.

Terdapat 3 tipe analisis bivariate:

- **Categorical - Categorical**
    - Chi Square
    - Bar Chart, 2-Y axis plot

- **Numerical - Numerical**
    - Correlation
    - Scatter Plot

- **Categorical - Numerical**
    - Ztest, ttest, ANOVA
    - Bar and Line Chart, 2-Y axis plot
