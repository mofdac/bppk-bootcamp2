### PandasGUI
PandasGUI is suitable for in-depth analysis with manual drag-and-drop capability.

```python
import pandas as pd
df = pd.read_csv('dataset/titanic.csv')

from pandasgui import show
show(df)
```

### dtale
D-Tale is the combination of a Flask back-end and a React front-end to bring you an easy way to view & analyze Pandas data structures.


```python
import dtale
import pandas as pd
df = pd.read_csv('dataset/titanic.csv')
d = dtale.show(df)
d.open_browser()
```

### Pandas-profilling
Pandas Profiling suitable for a quick analysis generation for a single variable.


```python
# Pandas profiling
import pandas as pd
df = pd.read_csv('dataset/titanic.csv')

from pandas_profiling import ProfileReport
prof = ProfileReport(df)
prof.to_file(output_file='output.html')
```

### SweetViz
Sweetviz suitable for analysis between dataset and to the target feature.

```python
## Sweet Viz
import sweetviz as sv

my_report = sv.analyze(df)
my_report.show_html() 
```

### Dataprep

```python
from dataprep.eda import create_report
df = pd.read_csv('dataset/titanic.csv')
create_report(df).show()
```
