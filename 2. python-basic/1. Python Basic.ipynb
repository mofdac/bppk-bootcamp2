### List

List dalam bahasa Python adalah daftar nilai yang dipisahkan koma (item) antara tanda kurung siku. Hal penting tentang daftar adalah item dalam list tidak harus sama jenisnya.

Anggota sebuah list diidentifikasi menggunakan index, index sendiri dimulai dari 0.

Contoh List:
```python
list1 = ['SETJEN', 'DJP', 1993, 2017]
list2 = [1, 2, 3, 4, 5 ]
list3 = ["a", "b", "c", "d"]
```


Terdapat beberapa method yang umum berkaitan dengan list

- ```len(list)```     : Mengetahui panjang list 

- ```append(item)```  : Menambahkan sebuah item ke dalam list

- ```extend(list)```  : Mengisi anggota list dengan anggota dari list lain

- ```remove(item)```  : Menghapus sebuah item ke dalam list

- ```count(item)```   : Menghitung keberadaan sebuah item pada sebuah list

- ```index(item)```   : Mengetahui posisi sebuah item pada sebuah list

- ```sort()```  : Mengurutkan item-item dalam list

- ```clear()```   : Mengosongkan nilai list

- ```pop(index)```    : Menghapus anggota list pada posisi tertentu

- ```reverse()``` : Membalikan urutan anggota list

## Conditional Statement (If...elif....else...)

Digunakan untuk mengendalikan kode yang akan dieksekusi dengan memperhatikan pemenuhan terhadap kondisi tertentu.

Syntax:

```python

if condition1:
    # baris dieksekusi jika condition1 terpenuhi
elif condition2:
    # baris dieksekusi jika condition2 terpenuhi
else:
    # baris dieksekusi jika condition1 & condition2 tidak terpenuhi


```

Jika kode kita membutuhkan logika percabangan sederhana, kita bisa menggunakan ternary if. Syntaxnya sebagai berikut:

```python

   c = a if a < b else b
    
   a, b = 10, 20
  
   print("Both a and b are equal" if a == b else "a is not equal to b") 

```

````python
condition1 = True
condition2 = True

if condition1:
    # baris dieksekusi jika condition1 terpenuhi
    pass
elif condition2:
    # baris dieksekusi jika condition2 terpenuhi
    pass
else:
    # baris dieksekusi jika condition1 & condition2 tidak terpenuhi
    pass
    
# ternary if
a, b = 10, 20
print("Both a and b are equal" if a == b else "a is not equal to b") 
````

# Variable Scope

1. A variable created in the main body of the Python code is a global variable and belongs to the global scope.

```python
    x = 300
    def fungsi():
      print(x)

```

2. A variable created inside a function is available inside that function:

```python
   def cek_kelulusan(nilai):   
       lulus = nilai > 50
       return lulus
    
   cek_kelulusan()
```

3. The local variable can be accessed from a function within the function:


```python
   def cek_kelulusan(nilai):   
       nilai = nilai * 100
       def cek_grade():
            if nilai > 30:
                return True
            else:
                return False
       return cek_grade()
    
   cek_kelulusan(2)
```
4. If you need to create a global variable, but are stuck in the local scope, you can use the global keyword.

```python

def myfunc():
  global x
  x = 300

myfunc()
print(x)

```

## Break, Continue dan Pass

- Pass jika dipanggil tidak akan menjalankan apa-apa. Biasanya digunakan jika kita ingin membuat fungsi tapi belum mendefinisikan proses di dalamnya

```python
def foo():
    pass

```
- Continue digunakan untuk menghiraukan blok kode setelahnya dan melanjutkan proses iterasi

```python
for i in range(1, 10):
    if i == 5:
        continue
    else:
        print(i)
```
- Break digunakan untuk menghentikan eksekusi kode pada sebuah iterasi
```python
for i in range(1, 10):
    if i == 5:
        break
    else:
        print(i)
```
