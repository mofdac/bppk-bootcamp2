## What is Matplotlib?
Matplotlib merupakan salah satu library di python yang biasa digunakan untuk keperluan visualisasi 2D. Data yang digunakan dapat juga berasal dari Pandas atau NumPy


### Matplotlib Pyplot
matplotlib.pyplot adalah kumpulan fungsi perintah. Setiap fungsi plot membuat beberapa perubahan pada visualisasi yang ditampilkan. Misalnya membuat x-axis dan y-axis, membuat label, dll.

Untuk melihat contoh-contoh visualisasi menggunakan (+soure codenya):https://matplotlib.org/stable/gallery/index.html

## Anatomy of Plot


(by Benjamin Root)

The Figure is the top-level container in this hierarchy. It is the overall window/page that everything is drawn on. You can have multiple independent figures and Figures can contain multiple Axes.

Most plotting ocurs on an Axes. The axes is effectively the area that we plot data on and any ticks/labels/etc associated with it. Usually we'll set up an Axes with a call to subplot (which places Axes on a regular grid), so in most cases, Axes and Subplot are synonymous.

Each Axes has an XAxis and a YAxis. These contain the ticks, tick locations, labels, etc.

### Contoh Plot Sederhana

```python
import matplotlib.pyplot as plt #Import PyPlot
```

```python
xdata = [1,2,3,4] # Data sumbu X
ydata = [1,2,3,4] # Data sumbu Y

xdata2 = [2,4,6,8] # Data sumbu X
ydata2 = [1,2,3,4] # Data sumbu Y

#plt.plot
plt.plot(xdata, ydata, color='red') #Plot 1
plt.plot(xdata2, ydata2, color='blue') #Plot 2

plt.xlabel('label x')  # Label sumbu X
plt.ylabel('label y') # Label sumbu Y
plt.title('Judul Plot') #judul plot
plt.show()
```

untuk mencari sebaran data

```python
# create color dictionary
colors = {'setosa':'red', 'versicolor':'green', 'virginica':'blue'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(iris['sepal length (cm)'])):
    ax.scatter(iris['sepal length (cm)'][i], iris['sepal width (cm)'][i],color=colors[iris['species'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepal width (cm)')
```

### Bar Plot

```python
# create a figure and axis 
fig, ax = plt.subplots() 

# count the occurrence of each class 
data = iris['species'].value_counts() 

# get x and y data 
points = data.index 
frequency = data.values 

# create bar chart 
ax.bar(points, frequency)

# set title and labels 
ax.set_title('Jumlah Spesies') 
ax.set_xlabel('Spesies') 
ax.set_ylabel('Jumlah')
```

### Histogram
